```python
df=df.drop(['link','id'],axis='columns')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.Make=le.fit_transform(df.Make)
df.Drive=le.fit_transform(df.Drive)
df[['Drive','Make']]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=df[["Battery Capacity","Top Speed","Fastcharge Speed","Total Power","Total Torque","Charge Speed","Width","Height","Acceleration 0 - 100 km/h","Seats","Max. Payload","Gross Vehicle Weight (GVWR)"]]
y=df["Electric Range"]
x=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
xtra=ExtraTreesRegressor(criterion="absolute_error")
xtra.fit(x_train,y_train)
y_pred=xtra.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)

```
